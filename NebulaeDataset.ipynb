```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.DataFrame({'Nebulae': np.random.rand(500),
                    'Andromeda': np.random.rand(500) * 2 + 28,
                    'Milky Way': np.random.rand(500) * 2 + 34,
                     'Black Holes': np.random.rand(500) * 2 + 40
                    })

plt.scatter(data['Nebulae'], data['Andromeda'])
plt.xlabel('Nebulae')
plt.ylabel('Andromeda')
plt.title('Nebulae vs. Andromeda')
plt.show()
```
